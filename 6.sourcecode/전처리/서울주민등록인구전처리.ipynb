```python
# 위에서 몇개의 행을 삭제하면서 인덱스가 연속적이지 않음
# index가 연속되도록 재설정(1~500)
pop.index = range(1,len(pop.index)+1)
pop
```

```python
# pop 데이터프레임에서 컬럼명만 추출
pop.columns

# 반복문(for문)을 활용하여 원하는 열(2017총인구 ~ 2019여자인구)만 속성값을 object -> int 형변환 
for column in pop.columns[2:]:
    pop[column] = pop[column].astype('int')
pop.info()
```

```python
pop_new1 = pd.DataFrame()
```

```python
# 행정구역(구별)
gu = []

# 연령대
ages = []

for i in pop.index[::2]:
#     print(i)
    gu.append(pop['행정구역(구별)'][i])
    ages.append(pop['연령대'][i])
```

```python
pop_new1['행정구역별(구별)'] = gu
pop_new1['연령대'] = ages
```

```python
pop_new1
```

```python
# 구별로 20개씩 행이 반복된다
# 20개씩 잘라서 계산 실행해서 새로운 데이터 프레임에 추가
pop_new2 = pd.DataFrame()
for i in pop.index[:250]:
    row = 2*i
    sum_ = pop.loc[row-1,'2017총인구':'2019여자인구']+pop.loc[row,'2017총인구':'2019여자인구']
    sum_.astype('int')
    pop_new2 = pop_new2.append(sum_,ignore_index=True)
```

```python
pop_new2
```

```python
pop_new = pd.concat([pop_new1,pop_new2],axis=1)
pop_new
```
